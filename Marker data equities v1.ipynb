##Equity spot prices


```python
SpotData = yf.download(tickers='hdfcbank.ns',start=date(2024,7,10) , end = date.today(),auto_adjust=False)
SpotData
```

```python
SpotData['Adj Close']
```

```python
SpotData.loc[:,'Adj Close']
```

```python
SpotData['Adj Close'].plot(figsize=(14,6),legend=True)
```

```python
SpotData=yf.download(tickers='HDFCBANK.NS',start=date(2025,1,1),end=date.today(),interval='1m')
SpotData
```

```python
Spottickers=['HDFCBANK.NS','AXISBANK.NS',]
SpotData=yf.download(tickers=Spottickers,start=date(2025,10,1),end=date.today(),auto_adjust=False,interval='1m')
SpotData
```

```python
SpotData['Adj Close'].plot(figsize=(14,7),legend=True)
```
