```python
import pandas as pd
import numpy as np
from scipy import stats
import yfinance as yf
from datetime import date
import matplotlib.pyplot as plt
```

```python
##Extarct Data
```

```python
SpotData = yf.download(tickers='^NSEI',start=date(2020,1,1),end=date.today(),auto_adjust=False,multi_level_index=False)['Adj Close'].fillna(method='bfill')
SpotReturns = SpotData.pct_change().iloc[1:]
```

```python
plt.figure(figsize = (14,6))
plt.scatter(x=SpotReturns.index , y=SpotReturns.values, label='Historcal Returns')
plt.legend()
plt.show()
            
```
